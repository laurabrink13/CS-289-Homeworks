Part B:

$$f(x) = \frac{1}{2||w||_1} * cos(<w, x>) \text{ for some } w \neq 0$$

```python
X1=np.arange(0,1,0.1)
X2=np.arange(0,1,0.1)

wl=[(1,1),(1,-1),(1,0),(-100,1),(5,50),(-50,-50)]

for j in range(0,len(wl)):
    w=wl[j]

    model_likelihood=np.zeros((len(X1),len(X2)))
    norm_w=sum([abs(x) for x in w])

    for ind1 in range(0,len(X1)):
        for ind2 in range(0,len(X2)):
            model_likelihood[ind1,ind2]=1/(2*norm_w)*np.cos(w[0]*X1[ind1]+w[1]*X2[ind2])


    L,M=np.meshgrid(X1, X2)
    CS = plt.contour(L, M, model_likelihood)
    plt.figure()
```
